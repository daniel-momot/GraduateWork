```python
import matplotlib.pyplot as plt
import numpy

Alg1 = list()
Alg2 = list()
MyAlg = list()

def do_process(rows):
    lines = list()
    sum1 = [0, 0]
    for i in range(len(rows)):
        tmp = rows[i]
        rows[i] = [x * 10000 for x in tmp]
        # секунды -> микросекунды ( * 1000000), количество повторных вызовов ( / 100)
        line = numpy.polyfit(x, rows[i], 1)
        lines.append(line)
        sum1[0] += line[0]
        sum1[1] += line[1]
    sum1[0] /= len(lines)
    sum1[1] /= len(lines)
    return (sum1[0], sum1[1])
```
